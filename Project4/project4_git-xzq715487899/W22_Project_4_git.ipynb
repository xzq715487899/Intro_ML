```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/aps1070-2019/datasets/master/LR_data.csv" , skipinitialspace=True)
```

```python
df
```

```python
# 1
X = df.loc[:,df.columns !='y']
y = df['y']
X=X.to_numpy()
y=y.to_numpy()
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2, random_state = 1)
print("shapes",X_train.shape,X_val.shape,y_train.shape,y_val.shape)
```

```python
def rmse(yPred, y):
    return np.sqrt(mean_squared_error(yPred, y))
```

```python
# 2
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_val_std = scaler.fit_transform(X_val)
reg = LinearRegression().fit(X_train_std, y_train)
yptrain = reg.predict(X_train_std)
ypval = reg.predict(X_val_std)
rmset = rmse(yptrain,y_train)
rmsev = rmse(ypval,y_val)
print("rmse train = ",rmset,"rmse val=" ,rmsev)
```

```python
# 3
poly = PolynomialFeatures(8)
X8t = poly.fit_transform(X_train)
X8V = poly.fit_transform(X_val)
X8t_std = scaler.fit_transform(X8t)
X8V_std = scaler.fit_transform(X8V)
reg8 = LinearRegression().fit(X8t_std, y_train)
yptrain8 = reg8.predict(X8t_std)
ypval8 = reg8.predict(X8V_std)
rmset8 = rmse(yptrain8,y_train)
rmsev8 = rmse(ypval8,y_val)
print("rmse train = ",rmset8,"rmse val=" ,rmsev8)
```

```python
# 4
lrl = []
rmsetl = []
rmsevl = []
for i in range(13):
  lr = 0.01 * 10**i
  clf = Ridge(alpha=lr)
  clf.fit(X8t_std, y_train)
  yptrain8r = clf.predict(X8t_std)
  ypval8r = clf.predict(X8V_std)
  rmset8r = rmse(yptrain8r,y_train)
  rmsev8r = rmse(ypval8r,y_val)
  lrl.append(lr)
  rmsetl.append(rmset8r)
  rmsevl.append(rmsev8r)
```

```python
summary = np.array((rmsetl,rmsevl,lrl))
table = pd.DataFrame(summary)
table.index = ['RMSEtrain','RMSEValication','Alpha']
table
```
